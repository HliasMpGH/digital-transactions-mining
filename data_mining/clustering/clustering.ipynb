Let's calculate the necessary metrics for each user and start formulating our clustering input

At first we'll calculate the total number of sales, money earned and quantity sold for each seller. We' ll also calculate the number of categories that were sold by each seller, the average sale price, the days that have passed since their last sale (final date of dataset used for this calculation) and a metric measuring the frequency of their sales during the time period in which they were active.

```python
max_date = df['full_date'].max()
```

```python
# Aggregating seller metrics
seller_metrics = df.groupby('seller_address').agg(
    total_sales=('seller_address', 'count'),
    total_money_earned=('wei_price', 'sum'),
    total_quantity_sold=('quantity', 'sum'),
    category_breadth = ('asset_category', 'nunique'),
    first_sale_date=('full_date', 'min'),
    last_sale_date=('full_date', 'max')
    
).reset_index()
# Calculate average sale price
seller_metrics['avg_sale_price'] = seller_metrics['total_money_earned'] / seller_metrics['total_sales']
# Calculate recency (days since last transaction)
seller_metrics['days_since_last_sale'] = (max_date - seller_metrics['last_sale_date']).dt.days
# Calculate another frequency metric (number of sales per days active)
seller_metrics['sales_frequency'] = seller_metrics['total_sales'] / (
    (seller_metrics['last_sale_date'] - seller_metrics['first_sale_date']).dt.days + 1)


seller_metrics
```

Moving on, we'll do the same for each buyer

```python
# Aggregating buyer metrics
buyer_metrics = df.groupby('buyer_address').agg(
    total_purchases=('buyer_address', 'count'),
    total_money_spent=('wei_price', 'sum'),
    total_quantity_bought=('quantity', 'sum'),
    category_breadth=('asset_category', 'nunique'),
    first_purchase_date=('full_date', 'min'),
    last_purchase_date=('full_date', 'max')
).reset_index()

# Calculate average purchase price
buyer_metrics['avg_purchase_price'] = buyer_metrics['total_money_spent'] / buyer_metrics['total_purchases']
# Calculate recency (days since last purchase)
buyer_metrics['days_since_last_purchase'] = (max_date - buyer_metrics['last_purchase_date']).dt.days
# Calculate another frequency metric (number of sales per days active)
buyer_metrics['purchase_frequency'] = buyer_metrics['total_purchases'] / (
    (buyer_metrics['last_purchase_date'] - buyer_metrics['first_purchase_date']).dt.days + 1
)

buyer_metrics
```

Now we are going to create variables to keep track of each sellers sales per category.

```python
# Aggregating sales by category
sales_by_category = df.groupby(['seller_address', 'asset_category']).size().unstack(fill_value=0)
sales_by_category = sales_by_category.reset_index()
# Remove index name
sales_by_category = sales_by_category.rename_axis(None, axis=1)
sales_by_category
```

Now we'll do the same for the purchases of each buyer

```python
# Aggregating purchases by category
purchases_by_category = df.groupby(['buyer_address', 'asset_category']).size().unstack(fill_value=0)
purchases_by_category = purchases_by_category.reset_index()
# Remove index name
purchases_by_category = purchases_by_category.rename_axis(None, axis=1)
purchases_by_category
```

Now let's merge the dataframes necessary to finalize the input for both the sellers and the buyers clustering.

```python
# Create seller clustering input
seller_metrics = seller_metrics.merge(sales_by_category, on= 'seller_address', how= 'inner')
seller_metrics
```

```python
# Create buyer clustering input
buyer_metrics = buyer_metrics.merge(purchases_by_category, on= 'buyer_address', how= 'inner')
buyer_metrics
```

Of course, given the fact that most users have been involved both in sales and purchases, many of the user addresses will be taken into consideration during both clustering implementations. However, the characteristics used each time differ, in order to better reflect one's behavior as a seller and as a buyer.

Let's export our files.

```python
seller_metrics.to_csv("seller_clustering_input.csv", index= False)
buyer_metrics.to_csv("buyer_clustering_input.csv", index= False)
```

Now to complete our clustering input, let's calculate the average holding period of an asset for each user. The holding period of an asset can be defined as the number of days that elapsed between the user's buying and selling of said asset. For assets that have no sales data attached to them, the final day of our dataset will be used to calculate this metric. For assets that have no purchase data attached to them, the first day of our dataset will be used.

```python
# Step 1: Determine dataset date range
min_date = df['full_date'].min()
max_date = df['full_date'].max()

# Step 2: Purchases (buyers acquiring assets)
purchases = df[['buyer_address', 'asset_id', 'full_date']].rename(columns={'buyer_address': 'user_address', 'full_date': 'purchase_date'})
purchases
```

```python
# Step 3: Sales (sellers selling assets)
sales = df[['seller_address', 'asset_id', 'full_date']].rename(columns={'seller_address': 'user_address', 'full_date': 'sale_date'})
sales
```

```python
# Step 4: Identify sellers without purchases
sellers_without_purchases = sales[~sales['user_address'].isin(purchases['user_address'])]
sellers_without_purchases['purchase_date'] = min_date  # Assume they acquired assets before dataset start

# Combine these sellers into the purchases dataframe
purchases = pd.concat([purchases, sellers_without_purchases[['user_address', 'asset_id', 'purchase_date']]])
purchases
```

## Seller clustering results

After many clustering attempts, the one that produces the best results is the one regarding the sellers. The results of the clustering will be analyzed in this section.

To find the optimal clustering the following process was conducted using RapidMiner. K-means was selected as the clustering method and different values where tested, as far as k and numerical measures are concerned.

The best performing clustering model is the one using k = 4 clusters and Kernel Euclidian Distance. This model also happens to produce the most meaningful results and so it is the model chosen for our analysis.

The results can be summarized as follows:
- Cluster 0 (Blue) - Specialized sellers (Εξειδικευμένοι πωλητές): This cluster has the highest average sale price and total earnings. They sell small quantities of assets that belong to specific category groups. Their total number of transactions is below average while their recency is average. They have a small number of sales that spans over a short time period and this is why they appear to have the largest frequency. To sum it up, they focus on a small amount of high value sales and sell assets that belong to particular categories they are heavily interested in. This cluster includes 55.83% of all sellers.
- Cluster 1 (Green) - Microsellers (Μικροπωλητές)  : This cluster contains sellers with low total earnings and average sale price that sell assets of various categories. Also, they have few transactions and have sold a small quantity of assets. However, they appear to have a decent recency. All in all, this cluster contains occasional sellers that rarely sell small quantities of assets for low prices without a particular focus in any category. This cluster includes 30.68% of all sellers.
- Cluster 2 (Yellow) - Bulk Traders (Χονδρέμποροι/Έμποροι χονδρικής) : This cluster contains sellers that have the largest amount of total sales, selling the highest quantity of assets. They appear to sell assets of a very wide variety of categories. Despite their heavy involvement in transactional activity, their average price per sale and their total earnings appear to be relatively low. They also have the best recency among clusters. To sum up, this cluster is characterized by the aforementioned heavy involvement in transactional activity regarding all sorts of low value and high quantity assets. This cluster includes 10.27% of all sellers.
- Cluster 3 (Red) - Dormant Sellers (Ανενεργοί/αδρανείς πωλητές) : This cluster contains sellers that are below average in every single category. They have the fewest transactions with low monetary gains. What truly sets them apart is their abysmal recency. The sellers of this cluster most probably just happened to sell a few random assets that were in their possession, and since haven't been at all active inside the marketplace. This cluster includes 3.22% of all sellers.
