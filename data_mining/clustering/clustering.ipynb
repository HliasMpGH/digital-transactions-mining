It appears that the majority of addresses found in seller id are also included in the buyer ids. However, to make sure no address is lost, when merging sellers and buyers data a full outer join must be conducted.

Let's calculate the necessary metrics for each user and start formulating our clustering input

At first we'll calculate the total number of sales, money earned and quantity sold for each seller.

```python
# Aggregating seller metrics
seller_metrics = df.groupby('seller_address').agg(
    total_sales=('seller_address', 'count'),
    total_money_earned=('wei_price', 'sum'),
    total_quantity_sold=('quantity', 'sum')
).rename_axis('user_address').reset_index()

seller_metrics
```

Moving on, we'll do the same for each buyer

```python
# Aggregating buyer metrics
buyer_metrics = df.groupby('buyer_address').agg(
    total_purchases=('buyer_address', 'count'),
    total_money_spent=('wei_price', 'sum'),
    total_quantity_bought=('quantity', 'sum')
).rename_axis('user_address').reset_index()

buyer_metrics
```

Now we are going to create variables to keep track of each sellers sales per category.

```python
# Aggregating sales by category
sales_by_category = df.groupby(['seller_address', 'asset_category']).size().unstack(fill_value=0)
sales_by_category = sales_by_category.rename_axis('user_address').reset_index()
# Remove index name
sales_by_category = sales_by_category.rename_axis(None, axis=1)
sales_by_category
```

Now we'll do the same for the purchases of each buyer

```python
# Aggregating purchases by category
purchases_by_category = df.groupby(['buyer_address', 'asset_category']).size().unstack(fill_value=0)
purchases_by_category = purchases_by_category.rename_axis('user_address').reset_index()
# Remove index name
purchases_by_category = purchases_by_category.rename_axis(None, axis=1)
purchases_by_category
```

Now it's time to merge the metrics we've calculated, we fill the NAs that come up, setting them equal to 0. Also, column names are renamed where needed to better differentiate between sales and purchases.

```python
# Merging all metrics
user_metrics = pd.merge(seller_metrics, buyer_metrics, on='user_address', how='outer').fillna(0)
user_metrics = pd.merge(user_metrics, sales_by_category, on='user_address', how='left').fillna(0)

# Rename sales category columns
sales_category_columns = sales_by_category.columns.difference(['user_address'])
user_metrics = user_metrics.rename(columns={col: f"sales_{col}" for col in sales_category_columns})

user_metrics = pd.merge(user_metrics, purchases_by_category, on='user_address', how='left').fillna(0)
# Rename purchases category columns
purchases_category_columns = purchases_by_category.columns.difference(['user_address'])
user_metrics = user_metrics.rename(columns={col: f"purchases_{col}" for col in purchases_category_columns})

user_metrics
```

Everything seems to be fine and there appear to be no missing values

```python
user_metrics.isna().sum()
```

Now to complete our clustering input, let's calculate the average holding period of an asset for each user. The holding period of an asset can be defined as the number of days that elapsed between the user's buying and selling of said asset. For assets that have no sales data attached to them, the final day of our dataset will be used to calculate this metric. For assets that have no purchase data attached to them, the first day of our dataset will be used.

```python
# Step 1: Determine dataset date range
min_date = df['full_date'].min()
max_date = df['full_date'].max()

# Step 2: Purchases (buyers acquiring assets)
purchases = df[['buyer_address', 'asset_id', 'full_date']].rename(columns={'buyer_address': 'user_address', 'full_date': 'purchase_date'})
purchases
```

```python
# Step 3: Sales (sellers selling assets)
sales = df[['seller_address', 'asset_id', 'full_date']].rename(columns={'seller_address': 'user_address', 'full_date': 'sale_date'})
sales
```

```python
# Step 4: Identify sellers without purchases
sellers_without_purchases = sales[~sales['user_address'].isin(purchases['user_address'])]
sellers_without_purchases['purchase_date'] = min_date  # Assume they acquired assets before dataset start

# Combine these sellers into the purchases dataframe
purchases = pd.concat([purchases, sellers_without_purchases[['user_address', 'asset_id', 'purchase_date']]])
purchases
```

All the addresses already exist on the purchases table so we can simply left join

```python
# Step 5: Merge purchases and sales
holdings = pd.merge(purchases, sales, on=['user_address', 'asset_id'], how='left')
holdings
```

```python
# Step 6: Assign holding periods
holdings['sale_date'] = holdings['sale_date'].fillna(max_date)  # Use max_date if no sale date exists
holdings['holding_period'] = (holdings['sale_date'] - holdings['purchase_date']).dt.days
holdings
```

```python
# Step 7: Calculate average holding period for each user
average_holding_period = holdings.groupby('user_address').agg(
    avg_holding_period=('holding_period', 'mean')
).reset_index()
average_holding_period
```

```python
# Step 8: Merge average holding period into the main user_metrics dataframe
user_metrics = pd.merge(user_metrics, average_holding_period, on='user_address', how='left').fillna(0)
user_metrics
```

```python
user_metrics.isna().sum()
```

There's an obvious problem because of multiple duplicate combinations of addresses and asset ids. That means that most probably average holding period is useless and should not be used for clustering users.
